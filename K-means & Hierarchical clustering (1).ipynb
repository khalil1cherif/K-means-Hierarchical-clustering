```python
import pandas as pd
data=pd.read_csv("CC GENERAL.csv")
```

```python
data
```

```python
data = data.drop(['CUST_ID'], axis=1)
data
```

```python
data.isnull().sum()
```

```python
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].median(), inplace = True)
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].median(), inplace = True)

```

```python
data.isnull().sum()
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative.head()
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['CASH_ADVANCE_TRX'] , data ["PAYMENTS"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("CASH_ADVANCE_TRX")
ax.set_ylabel("PAYMENTS")
plt.colorbar(scatter)
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data, method="complete"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(data)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(data)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
data["label"]=agglomerative
```

```python
plt.scatter(data["CASH_ADVANCE_TRX"][data.label == 0],          
            data["PAYMENTS"][data.label == 0],s=80,c='magenta',label='Careful')
plt.scatter(data["CASH_ADVANCE_TRX"][data.label == 1],
           data["PAYMENTS"][data.label == 1],s=80,c='yellow',label='Standard')
plt.scatter(data["CASH_ADVANCE_TRX"][data.label == 2],
           data["PAYMENTS"][data.label == 2],s=80,c='green',label='Target')
plt.scatter(data["CASH_ADVANCE_TRX"][data.label == 3], 
           data["PAYMENTS"][data.label == 3],s=80,c='cyan',label='Careless')
plt.scatter(data["CASH_ADVANCE_TRX"][data.label == 4], 
           data["PAYMENTS"][data.label == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('CASH_ADVANCE_TRX')
plt.ylabel('PAYMENTS')
plt.legend()
plt.show()
```
